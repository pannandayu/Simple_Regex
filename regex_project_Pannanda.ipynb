**Log Data**

extract host, user name, time, request and convert into dictionary

```python
import re


def logs():
    logs = []
    pattern = '\
    (?P<host>(\d+\.){3}\d+)\
    (\s-\s)\
    (?P<user_name>\S+)\
    (\s)\
    \[(?P<time>[-+\w\s:/]+)\]\
    (\s)\
    "(?P<request>\S.*)"'

    with open("logdata.txt", "r") as file:
        logdata = file.read()
    for items in re.finditer(pattern, logdata, re.VERBOSE):
        logs.append(items.groupdict())
    return logs

logs()
```
